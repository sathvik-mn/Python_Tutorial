```python
#Simple Linear Reg
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

from pylab import rcParams
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 10, 8
```

```python
rooms = 2*np.random.randn(100,1)+3
rooms [0:10]
```

```python
price = 265 + 6*rooms + abs(np.random.randn(100,1))
price[0:10]
```

```python
plt.plot(rooms,price, 'r^' )
plt.xlabel('# of rooms, 2024 Average')
plt.ylabel('2024 Average home price, $1000s USD')
plt.show()
```

```python
#Rooms as predictor
x= rooms
y = price

LinReg = LinearRegression()
LinReg.fit(x, y)
print(LinReg.intercept_, LinReg.coef_)
```

```python
print(LinReg.score(x,y))
```

```python
#score returns coef of determination, which is the R^2 of the prediction.
#Model is permofming well
```
